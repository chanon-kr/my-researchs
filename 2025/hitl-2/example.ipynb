```python
from dotenv import load_dotenv
load_dotenv()

from graph.graph import graph
from langgraph.types import Command

thread = {"configurable": {"thread_id": "4"}}

r= graph.invoke(
                {'messages' : 'What is the meaning of life'}, 
                # {'messages' : 'What is 1 + 1'}, 
                config= thread)

while '__interrupt__' in r :
    hitl_msg= r['__interrupt__'][-1].value
    feedback= input(hitl_msg)
    r= graph.invoke(Command(resume=feedback), config=thread)
    
r
```
